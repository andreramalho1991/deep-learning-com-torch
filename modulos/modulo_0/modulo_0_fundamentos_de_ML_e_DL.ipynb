```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots(figsize=(10,7))
ax.axis('off')

plt.title('RelaÃ§Ã£o entre InteligÃªncia Artificial, Machine Learnig e Deep Learning')

ia = plt.Circle((0.5, 0.5), 0.4, color='orange', alpha =0.3)
ax.add_patch(ia)
plt.annotate('InteligÃªncia Artificial', xy=(0.1, 0.5), xytext=(0.1, 0.5))

ml = plt.Circle((0.6, 0.5), 0.3, color='red', alpha =0.3)
ax.add_patch(ml)
plt.annotate('Machine Learning', xy=(0.3, 0.5), xytext=(0.3, 0.5))

dl = plt.Circle((0.7, 0.5), 0.2, color='blue', alpha =0.3)
ax.add_patch(dl)
plt.annotate('Deep Learning', xy=(0.5, 0.5), xytext=(0.5, 0.5))

plt.show()
```
